```python
### Dependent and independent features
# X is independent variables isme exicted ke alawa sb row hoga
X = data.drop('Exited',axis=1)
Y = data['Exited']
```

```python
X.head()
```

```python
categorical_cols = X.select_dtypes(include= 'object').columns
numerical_cols = X.select_dtypes(exclude= 'object').columns


print("categorical column: ", categorical_cols)
print("numerical column: ", numerical_cols)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler

#pipeline
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
gender_cat = ['Female', 'Male']
geo_cat = ['France', 'Spain','Germany']
```

```python
num_pipeline = Pipeline(steps=[('scaler', StandardScaler())])

cat_pipeline = Pipeline(steps=[('ordinalencoder', OrdinalEncoder(
    categories=[geo_cat, gender_cat]))])

preprocessor = ColumnTransformer(
    [("num_pipeline", num_pipeline, numerical_cols),
     ("cat_pipeline", cat_pipeline, categorical_cols)]
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
# Applying the preprocessing pipelines to the training and testing data
X_train = pd.DataFrame(preprocessor.fit_transform(X_train), columns= preprocessor.get_feature_names_out())
X_test = pd.DataFrame(preprocessor.transform(X_test), columns= preprocessor.get_feature_names_out())
```

```python
X_train.head()
```

### Feature Engeineering

```python
import pickle
with open("Saved_model/preprocessor.pkl", "wb") as f:
    pickle.dump(preprocessor, f)
```

ANN IMPLMENTATIoN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
#comma denote dimension 
X_train.shape[1],
```

```python
# Hidden node ko create krne keliye Dense use krte hai
#trenable parameter means create 
#No. of nodes = no. of bias 

model = Sequential([
    Dense(64, activation='relu', input_shape = (X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate = 0.001)
```

```python
model.compile(
    optimizer = opt,
    loss = "binary_crossentropy",
    metrics = ["accuracy"]
)

```

```python
log_dir = "logs/"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
early_stopping_callback = EarlyStopping(
    monitor = "val_loss",
    patience = 5,
    restore_best_weights = True
)
```

#3 Train Model


```python
history = model.fit(
    X_train, Y_train,
    validation_data = (X_test, Y_test),
    epochs = 100,
    callbacks = [tensorflow_callback, early_stopping_callback]
)
```

```python
model.save("model.h5")
```

```python
loss = history.history['loss']
acc = history.history['accuracy']
val_loss = history.history['val_loss']
val_acc = history.history['val_accuracy']
```

```python
epochs = len(loss)
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
